The data is not really Gaussian, but skewing doesn't seem to be too severe. Let's setup the data, and make an attempt with a Naive Bayes with independent Gaussian distributions:

```python
train.strat_train_test_split(attrib='target', test_size=0.2)
train.get_labels(label_name='target')

train._train.drop("ID_code", axis=1, inplace=True)
train._test.drop("ID_code", axis=1, inplace=True)

train.set_attrib_types(cat_attribs=None)

pipeline = train.set_pipeline()

train_prepared = pipeline.fit_transform(train._train)
test_prepared = pipeline.transform(train._test)
train_labels = train._labelstrain
test_labels = train._labelstest
```

```python
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
clf.fit(train_prepared, train_labels)
train_predictions = clf.predict(train_prepared)
print('Validation AUC =',roc_auc_score(train_labels, train_predictions))
```

The score is not so much better than a completely random guess. The Naive-Bayes method relies also on the priors, which is something we haven't specified. Let's play with them to see what we get:

```python
for prior in np.arange(0.05, 0.95, 0.05):
    priors_clf = GaussianNB(priors=[prior, 1-prior])
    priors_clf.fit(train_prepared, train_labels)
    train_predictions = priors_clf.predict(train_prepared)
    print(f'Priors: {prior}, {1-prior}; Validation AUC =',roc_auc_score(train_labels, train_predictions))
```

A balanced prior yields the best results, with a precision of 81%. For a simple model, this is not bad at all. Let's apply the model to test:

```python
priors_clf = GaussianNB(priors=[0.5, 0.5])
priors_clf.fit(train_prepared, train_labels)
test_predictions = priors_clf.predict(test_prepared)
print('Validation AUC =',roc_auc_score(test_labels, test_predictions))
```

Not bad, the model doesn't overfit. This is expected, since a Naive Bayes model makes little assumptions about the data. Now, what can we do to improve? Let's try a different classifier: LGBMClassifier. Light GBM is a gradient boosting framework that uses tree based learning algorithm. Light GBM grows tree vertically while other algorithm grows trees horizontally meaning that Light GBM grows tree leaf-wise while other algorithm grows level-wise. It will choose the leaf with max delta loss to grow. When growing the same leaf, Leaf-wise algorithm can reduce more loss than a level-wise algorithm.

Here's an introductory article about it: https://medium.com/@pushkarmandot/https-medium-com-pushkarmandot-what-is-lightgbm-how-to-implement-it-how-to-fine-tune-the-parameters-60347819b7fc

```python



params = {
        'task': 'train',
        'objective': 'binary',
        'metric': 'binary_error',
        'boosting_type': 'dart',    
        'learning_rate': 0.1,       
        'max_depth': 8,             
        'min_data_in_leaf': 5,      
        'max_bin' : 300,            
        'num_leaves' : 128,         
        'verbose': 100,
        'lambda_l2': 1
}

lgb_train = lgbm.Dataset(train_prepared, label=train_labels, free_raw_data=False)
lgb_eval = lgbm.Dataset(test_prepared, label=test_labels, reference=lgb_train, free_raw_data=False)
clf = lgbm.train(params, lgb_train, valid_sets=[lgb_train,lgb_eval],
            valid_names=['train', 'eval'], num_boost_round=100)

train_prediction = clf.predict(train_prepared)
test_prediction = clf.predict(test_prepared)
```

```python
print('Training Validation AUC =',roc_auc_score(train_labels, train_prediction))
print('Testing Validation AUC =',roc_auc_score(test_labels, test_prediction))
```

Compared to the top Kaggle competitors, we are just 9% behind in our test. Notice that we are very succeptible to overfitting, even after large $\mathcal{l_2}$ regularization. What we can try to do is some sort of learning rate annealing, forcing the model to get out of the minima it is:

```python
print('Continue training')

gbm = lgbm.train(params,
                lgb_train,
                num_boost_round=100,
                init_model=clf,
                learning_rates=lambda iter: 0.1 * (0.99 ** iter),
                valid_sets=lgb_eval)
```

```python
train_prediction = gbm.predict(train_prepared)
test_prediction = gbm.predict(test_prepared)

print('Training Validation AUC =',roc_auc_score(train_labels, train_prediction))
print('Testing Validation AUC =',roc_auc_score(test_labels, test_prediction))
```

This is a 3% improvement, but the model is starting to reach its limit in fitting (Training Validation AUC close to 100%). Let's try to overcome this overfitting by splitting the training data into multiple folds, and training a LGBMClassifier for every fold. But first, let's perform a Bayesian optimization of the parameters, which goes through multiple parameter tunning like GridSearch:

```python
from bayes_opt import BayesianOptimization

def bayes_parameter_opt_lgb(X, y, init_round=15, opt_round=25, n_folds=5, 
                            random_seed=42, n_estimators=10000, learning_rate=0.05, 
                            output_process=False):
    # prepare data
    lgb_train = lgbm.Dataset(data=X, label=y, free_raw_data=False)

    # parameters
    def lgb_eval(num_leaves, feature_fraction, bagging_fraction, max_depth, lambda_l1, lambda_l2, min_split_gain, min_child_weight):
        params = {'application':'binary','num_boosting_round': n_estimators, 
                  'learning_rate':learning_rate, 'early_stopping_rounds':100, 'metric':'auc'}
        params["num_leaves"] = int(round(num_leaves))
        params['feature_fraction'] = max(min(feature_fraction, 1), 0)
        params['bagging_fraction'] = max(min(bagging_fraction, 1), 0)
        params['max_depth'] = int(round(max_depth))
        params['lambda_l1'] = max(lambda_l1, 0)
        params['lambda_l2'] = max(lambda_l2, 0)
        params['min_split_gain'] = min_split_gain
        params['min_child_weight'] = min_child_weight
        cv_result = lgbm.cv(params, lgb_train, nfold=n_folds, seed=random_seed, stratified=True, 
                            verbose_eval =200, metrics=['auc'])
        return max(cv_result['auc-mean'])
    # range 
    lgbBO = BayesianOptimization(lgb_eval, {'num_leaves': (45, 70),
                                            'feature_fraction': (0.1, 0.9),
                                            'bagging_fraction': (0.8, 1),
                                            'max_depth': (5, 8),
                                            'lambda_l1': (0, 5),
                                            'lambda_l2': (0, 3),
                                            'min_split_gain': (0.001, 0.1),
                                            'min_child_weight': (5, 50)}, random_state=0)
    # optimize
    lgbBO.maximize(init_points=init_round, n_iter=opt_round)
    
    # output optimization process
    if output_process==True: lgbBO.points_to_csv("bayes_opt_result.csv")
    
    # return best parameters
    return lgbBO.max

opt_params = bayes_parameter_opt_lgb(train_prepared, train_labels, init_round=5, opt_round=10, 
                                     n_folds=3, random_seed=42, n_estimators=100, learning_rate=0.05)
```

```python
params = {
    "objective" : "binary",
    "metric" : "auc",
    "boosting": 'gbdt',
    "max_depth" : 7,
    "num_leaves" : 63,
    "learning_rate" : 0.01,
    "bagging_freq": 5,
    "bagging_fraction" : 0.9,
    "feature_fraction" : 0.1,
    "min_split_gain": 0.025,
    "min_child_weight": 25,
    "min_data_in_leaf": 150,
    "min_sum_heassian_in_leaf": 10,
    "tree_learner": "serial",
    "boost_from_average": "false",
    "bagging_seed" : 42,
    "verbosity" : 1,
    "lambda_l1": 0.25,
    "lambda_l2": 1,
    "seed": 42}

features = [feat for feat in train.df.columns if feat not in ['ID_code', 'target']]

# LightGBM GBDT with KFold or Stratified KFold
# Parameters from Tilii kernel: https://www.kaggle.com/tilii7/olivier-lightgbm-parameters-by-bayesian-opt/code
def kfold_lgbm(num_folds, params, stratified=False, debug=False, seed=42):
    
    test_prediction = np.zeros(len(train.df))
    predictions = np.zeros(len(test.df))
    #getVal = np.zeros(len(train.df))
    feature_importance_df = pd.DataFrame()
    
    # Cross validation model
    if stratified:
        folds = StratifiedKFold(n_splits=num_folds, shuffle=False, random_state=seed)
    else:
        folds = KFold(n_splits=num_folds, shuffle=False, random_state=seed)

    
    for n_fold, (train_idx, valid_idx) in enumerate(folds.split(train.df.values, train.df.target.values),1):
        train_x, train_y = train.df.iloc[train_idx][features], train.df.target.iloc[train_idx]
        valid_x, valid_y = train.df.iloc[valid_idx][features], train.df.target[valid_idx]

        lgb_train = lgbm.Dataset(train_x, label=train_y)
        lgb_eval = lgbm.Dataset(valid_x, label=valid_y)
        
        gbm = lgbm.train(params, lgb_train,
                         valid_sets=[lgb_train,lgb_eval],
                         valid_names=['train', 'eval'],
                         num_boost_round=10000, verbose_eval=5000,
                         early_stopping_rounds = 2500)
        
        fold_importance_df = pd.DataFrame()
        fold_importance_df["feature"] = features
        fold_importance_df["importance"] = gbm.feature_importance()
        fold_importance_df["fold"] = n_fold
        feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)
        
        test_prediction[valid_idx] = gbm.predict(train.df.iloc[valid_idx][features], num_iteration=gbm.best_iteration)

        print(f'{n_fold}o. fold Validation AUC ={roc_auc_score(train.df.target[valid_idx], test_prediction[valid_idx])}\n')

        predictions += gbm.predict(test.df[features], num_iteration=gbm.best_iteration) / folds.n_splits
    
    print("CV score: {:<8.5f}".format(roc_auc_score(train.df.target, test_prediction)))
    
    cols = (feature_importance_df[["feature", "importance"]]
        .groupby("feature")
        .mean()
        .sort_values(by="importance", ascending=False)[:1000].index)
    best_features = feature_importance_df.loc[feature_importance_df.feature.isin(cols)]

    plt.figure(figsize=(14,26))
    sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance",ascending=False))
    plt.title('LightGBM Features (averaged over folds)')
    plt.tight_layout()
    plt.show()
    
    return predictions
```

```python
predictions = kfold_lgbm(num_folds=10, params=params, stratified=True)
```

While the model still overfits, we got a significant improvement, reaching almost 90%, close to Kaggle's leaderboard best (93%). While other heavy approaches, such as neural networks could be used, and even data augmentation, where we could copy existing data and shuffle inside the training set, this challenge was very useful to learn about the power of LGBM.
